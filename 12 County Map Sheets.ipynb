```python
# georeferencing a map sheet 

import subprocess
image_path = "image.png"
output_path = "output_georeferenced.png"
bounding_box = "53.1111876266 11.7303748998 55.5611387985 13.1036943879"
srs = "EPSG:4326"

command = f"gdal_translate {image_path} {output_path} -a_ullr {bounding_box} -a_srs {srs}"
subprocess.run(command, shell=True)
print(f"Output image saved to: {output_path}")
```

```python
# automated version of merging, georeferencing, and projecting each counties 16 individual map sheets into one raster layer

county_name = "west_sussex"

data = pd.read_csv(f'{county_name}_maps/metadata.csv')
new_df = data[['name', 'coordinates']].copy()
new_df['coordinates'] = new_df['coordinates'].str.replace(',', '').str.replace('(', '').str.replace(')', '')
new_df['name'] = new_df['name'].str.rstrip('.png')
#swap the coordinates so map is not upside down (order from 1234 to 1432)
new_df['coordinates'] = new_df['coordinates'].apply(lambda x: ' '.join([x.split()[0], x.split()[3], x.split()[2], x.split()[1]]))

os.mkdir(f"Georeferenced_maps/{county_name}")
srs = "EPSG:4326"
for index, row in new_df.iterrows():
    image_path = f"{county_name}_maps/{row['name']}.png"
    output_path = f"Georeferenced_Maps/{county_name}/{row['name']}_georeferenced.png"
    bounding_box = row['coordinates']
    command = f"gdal_translate {image_path} {output_path} -a_ullr {bounding_box} -a_srs {srs} -co TILED=YES"
    subprocess.run(command, shell=True)
    #maybe add:    command = None   like I did for one below
    print(f"Processed image: {image_path}")
    print(f"Output path: {output_path}")

# merge raster layers
georeferenced_folder = f"Georeferenced_Maps/{county_name}"
files_to_mosaic = [os.path.join(georeferenced_folder, file) for file in os.listdir(georeferenced_folder) if file.endswith("_georeferenced.png")]
g = gdal.Warp(f"merged/{county_name}.tif", files_to_mosaic, format="GTiff",
              options=["TILED=YES"]) # if you want
g = None # Close file and flush to disk

input_path = f"merged/{county_name}.tif"
output_path = f"merged/{county_name}_BNG.tif"
target_srs = "EPSG:27700"
# Open the input raster
input_ds = gdal.Open(input_path)
# Get the geotransform and projection of the input raster
geotransform = input_ds.GetGeoTransform()
projection = input_ds.GetProjection()
# Create a new raster with the target projection
target_ds = gdal.Warp(output_path, input_ds, dstSRS=target_srs)
# Close the input and target datasets
input_ds = None
target_ds = None
print(f"Projected raster saved to: {output_path}")

extent = rasterio.open(f'merged/{county_name}_BNG.tif')
print(extent.bounds)
```

##### classifier

```python
from mapreader import ClassifierContainer
my_classifier = ClassifierContainer(
    model="resnet18", dataloaders=dataloaders, labels_map={0: "no FBL", 1: "FBL"}
)
```

```python
my_classifier.labels_map
```

```python
my_classifier.model_summary()
```

##### Set up criterion, optimizer and scheduler

```python
my_classifier.add_criterion("cross-entropy")
params_to_optimize = my_classifier.generate_layerwise_lrs(
    min_lr=1e-4, max_lr=1e-3, spacing="geomspace"
)
my_classifier.initialize_optimizer(params2optimize=params_to_optimize)
my_classifier.initialize_scheduler()
```

##### Train/fine-tune model

```python
my_classifier.train(
    num_epochs=20,
    save_model_dir="./models_tutorial",
    tensorboard_path="tboard_tutorial",
    tmp_file_save_freq=2,
    remove_after_load=False,
    print_info_batch_freq=10,
)
```

```python
#list(my_classifier.metrics.keys())
my_classifier.plot_metric(
    y_axis=["epoch_loss_train", "epoch_loss_val"],
    y_label="Loss",
    legends=["Train", "Valid"],
    figsize=(7,2)
)
```
